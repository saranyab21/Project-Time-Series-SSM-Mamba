```python
# ========= 0) Imports & paths (project-structure aware) =========
import os, re, zipfile, warnings, glob
from pathlib import Path

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from tsfresh.feature_extraction import extract_features, EfficientFCParameters

# --- Paths based on your folder structure shown in screenshots ---
# This notebook lives in: .../gait-mamba/src
SRC_DIR   = Path.cwd()
ROOT_DIR  = SRC_DIR.parent                  # .../gait-mamba
DATA_DIR  = ROOT_DIR / "data"
RAW_DIR   = DATA_DIR / "raw" / "gait-in-parkinsons-disease-1.0.0"
PROC_DIR  = DATA_DIR / "processed"
FIG_DIR   = ROOT_DIR / "reports" / "figs"

for d in [RAW_DIR, PROC_DIR, FIG_DIR]:
    d.mkdir(parents=True, exist_ok=True)

# Try to find the ZIP in a few sensible places (optional; only used if RAW_DIR missing)
ZIP_CANDIDATES = [
    SRC_DIR / "gait-in-parkinsons-disease-1.0.0.zip",
    ROOT_DIR / "gait-in-parkinsons-disease-1.0.0.zip",
    DATA_DIR / "raw" / "gait-in-parkinsons-disease-1.0.0.zip",
    Path("gait-in-parkinsons-disease-1.0.0.zip"),
]
ZIP_PATH = next((z for z in ZIP_CANDIDATES if z.exists()), None)

SEED = 42

# ========= 1) Unpack zip (idempotent; only if raw folder not present) =========
if not RAW_DIR.exists():
    if ZIP_PATH is None:
        raise FileNotFoundError(
            "Raw folder not found and ZIP not located. "
            "Place 'gait-in-parkinsons-disease-1.0.0.zip' under project root or data/raw."
        )
    # Extract such that we end up with .../data/raw/gait-in-parkinsons-disease-1.0.0/*
    with zipfile.ZipFile(ZIP_PATH, "r") as z:
        # If the zip already contains the top-level folder, extract to data/raw
        # Otherwise, extract to a temp folder and move — but in practice PhysioNet zip has the folder.
        z.extractall(DATA_DIR / "raw")
    if not RAW_DIR.exists():
        # Fallback: maybe zip extracted flat; create the folder and move files in
        RAW_DIR.mkdir(parents=True, exist_ok=True)
        for p in (DATA_DIR / "raw").glob("*"):
            if p.is_file():
                p.rename(RAW_DIR / p.name)

print(f"Data at: {RAW_DIR}")

# ========= 2) Discover trials & parse ids =========
# Accept cohort prefixes seen in this dataset: GaCo, GaPt, JuCo, JuPt, SiCo, SiPt, SHA1/2, etc.
# Our regex parses: <prefix><two_digit_subject>_<two_digit_trial>.txt
TRIAL_RE = re.compile(r"^(?P<prefix>[A-Za-z]+)(?P<sid>\d{2})_(?P<tid>\d{2})\.txt$")

all_txt = sorted(glob.glob(str(RAW_DIR / "**" / "*.txt"), recursive=True))
names   = [os.path.basename(p) for p in all_txt]
matches = [(p, TRIAL_RE.match(os.path.basename(p))) for p in all_txt]
trial_paths = [(p, m) for p, m in matches if m is not None]

def prefix_to_label(prefix: str) -> int:
    up = prefix.upper()
    # Patients if *Pt* in prefix; Controls if *Co* in prefix; otherwise unknown (-1)
    if "PT" in up: return 1
    if "CO" in up: return 0
    return -1

trials = []
for path, m in trial_paths:
    pr  = m.group("prefix")
    sid = m.group("sid")
    tid = m.group("tid")
    label = prefix_to_label(pr)
    trials.append({
        "filepath": path,
        "filename": os.path.basename(path),
        "prefix": pr,
        "subject_id": sid,
        "trial_id": tid,
        "label_from_name": label
    })

trials = pd.DataFrame(trials).sort_values(["subject_id", "trial_id"]).reset_index(drop=True)
print(f"Total .txt files found: {len(all_txt)} | Recognized trial files: {len(trials)}")
if not trials.empty:
    print("Sample trials:", *trials["filename"].head(8).tolist(), sep="\n- ")

# ========= 3) Sampling rate detection =========
def detect_fs(raw_dir: Path) -> int:
    fmt = raw_dir / "format"
    if fmt.exists():
        txt = fmt.read_text(errors="ignore")
        m = re.search(r"(?i)(sampling\s*frequency|fs)\D+(\d+)", txt)
        if m:
            fs = int(m.group(2))
            print(f"Detected sampling rate from 'format': {fs} Hz")
            return fs
    warnings.warn("Could not detect sampling rate; defaulting to 100 Hz")
    return 100

FS = detect_fs(RAW_DIR)
win_samples  = max(1, round(FS * 0.030))  # ~30 ms
step_samples = max(1, round(FS * 0.015))  # ~15 ms
print(f"Sliding window: {win_samples} samples (~30 ms), step: {step_samples} samples (~15 ms) at {FS} Hz")

# ========= 4) Load demographics (robust; no deprecated pandas internals) =========
def load_demographics(raw_dir: Path) -> pd.DataFrame:
    # Prefer Excel, fallback to HTML file (PhysioNet sometimes ships both)
    xls  = raw_dir / "demographics.xls"
    html = raw_dir / "demographics"
    if xls.exists():
        df = pd.read_excel(xls)
    elif html.exists():
        # Load first table from HTML file on disk
        df = pd.read_html(html)[0]
    else:
        raise FileNotFoundError("No demographics.xls or demographics (html) found in RAW_DIR.")

    # Standardize column names
    df.columns = [str(c).strip().lower().replace(" ", "_") for c in df.columns]

    # Identify subject id column
    cand_sid = [c for c in df.columns if c in ("subject","subject_id","subj","patient","code","id")]
    if not cand_sid:
        raise ValueError(f"Cannot find subject id in demographics columns: {df.columns}")
    sid_col = cand_sid[0]

    out = pd.DataFrame()
    out["subject_id"] = (
        df[sid_col].astype(str)
          .str.extract(r"(\d+)")
          .iloc[:, 0].astype(int).astype(str).str.zfill(2)
    )

    # sex/gender
    sex_col = "sex" if "sex" in df.columns else ("gender" if "gender" in df.columns else None)
    out["sex"] = (df[sex_col] if sex_col else pd.Series(["unknown"] * len(out))).astype(str).str.lower()

    # age
    out["age"] = pd.to_numeric(df["age"], errors="coerce") if "age" in df.columns else np.nan

    # optional label (derive if absent)
    lbl_col = next((c for c in ("label","group","diagnosis","status") if c in df.columns), None)
    if lbl_col is not None:
        g = df[lbl_col].astype(str).str.upper()
        out["label"] = np.where(
            g.str.contains("PD"),
            1,
            np.where(g.str.contains("CO") | g.str.contains("CONTROL") | g.str.contains("HEALTH"), 0, np.nan),
        )
    else:
        out["label"] = np.nan

    return out.drop_duplicates(subset="subject_id").reset_index(drop=True)

demo = load_demographics(RAW_DIR)

# If demographics label is NaN, derive from filenames by majority vote per subject
if not trials.empty:
    name_label = trials.groupby("subject_id")["label_from_name"] \
                       .agg(lambda s: s[s != -1].mode().iloc[0] if (s != -1).any() else np.nan)
    demo = demo.merge(name_label.rename("label_name"), on="subject_id", how="left")
    demo["label"] = demo["label"].fillna(demo["label_name"]).astype(float)
    demo = demo.drop(columns=["label_name"], errors="ignore")

print("\nLoaded demographics columns:", list(demo.columns))
print("subjects:", demo["subject_id"].nunique())
print("label counts:", demo["label"].value_counts(dropna=False).to_dict())

# For Deliverable slide: patients vs controls, age, male/female
def describe_dataset(demo_df: pd.DataFrame):
    sex = demo_df["sex"].astype(str).str.lower().str.extract("(male|female)")[0].fillna("unknown")
    males   = int((sex == "male").sum())
    females = int((sex == "female").sum())
    ages = pd.to_numeric(demo_df["age"], errors="coerce")
    n_pd = int((demo_df["label"] == 1).sum())
    n_co = int((demo_df["label"] == 0).sum())
    print(f"\nDataset summary -> Patients: {n_pd} | Controls: {n_co} | Males: {males} | Females: {females}")
    if ages.notna().any():
        print(f"Age: mean={ages.mean():.1f}, std={ages.std():.1f}, min={ages.min():.0f}, max={ages.max():.0f}")

describe_dataset(demo)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ==== User inputs: set your file paths here ====
PD_FILE = "C:/Users/admin/OneDrive/Desktop/SEM-2-3/Project-Time Series/gait-mamba/data/raw/gait-in-parkinsons-disease-1.0.0/GaPt05_01.txt" #GaPt05
HC_FILE = "C:/Users/admin/OneDrive/Desktop/SEM-2-3/Project-Time Series/gait-mamba/data/raw/gait-in-parkinsons-disease-1.0.0/JuCo25_01.txt" #JuCo25

# ==== Parameters ====
FS = 100                    # sampling frequency (Hz) in GaitPDB
POINTS_PER_CYCLE = 101      # 0..100% stride
USE_SCIPY = True            # if scipy is installed, leave True; otherwise set False

# ------------------------------------------------
# Helpers
# ------------------------------------------------
def load_subject_vgrf(path, n_sensors_per_foot=8, sep=None):
    """
    Load a GaitPDB subject file and return left and right total vGRF.
    - The dataset has 8 sensors per foot (16 total columns).
    - Files are typically whitespace/tab delimited.
    """
    df = pd.read_csv(path, sep=sep, header=None)
    left = df.iloc[:, :n_sensors_per_foot].sum(axis=1).to_numpy(dtype=float)
    right = df.iloc[:, n_sensors_per_foot:2*n_sensors_per_foot].sum(axis=1).to_numpy(dtype=float)
    return left, right

def _find_peaks_scipy(signal, fs, min_stride_s=0.8):
    from scipy.signal import find_peaks
    # dynamic height threshold: between median and 75th percentile
    thr = 0.5*(np.median(signal) + np.percentile(signal, 75))
    dist = int(min_stride_s * fs)  # minimum distance between steps
    peaks, _ = find_peaks(signal, height=thr, distance=dist)
    return peaks

def _find_peaks_fallback(signal, fs, min_stride_s=0.8):
    # simple percentile threshold + local max window
    thr = 0.5*(np.median(signal) + np.percentile(signal, 75))
    dist = int(min_stride_s * fs)
    cand = np.where(signal > thr)[0]
    if cand.size == 0:
        return np.array([], dtype=int)
    # group contiguous indices; take max in each group
    groups = np.split(cand, np.where(np.diff(cand) > 1)[0] + 1)
    peaks = []
    last_idx = -np.inf
    for g in groups:
        if g.size == 0:
            continue
        idx = g[np.argmax(signal[g])]
        if idx - last_idx >= dist:
            peaks.append(idx)
            last_idx = idx
    return np.array(peaks, dtype=int)

def find_stride_peaks(signal, fs=FS, min_stride_s=0.8):
    if USE_SCIPY:
        try:
            return _find_peaks_scipy(signal, fs, min_stride_s)
        except Exception:
            pass
    return _find_peaks_fallback(signal, fs, min_stride_s)

def normalize_cycles(signal, peak_idx, points=POINTS_PER_CYCLE):
    """
    Given a signal and indices of consecutive peaks (same foot),
    slice cycles [peak_i, peak_{i+1}) and resample each to 'points'.
    Returns array shape (n_cycles, points).
    """
    cycles = []
    for i in range(len(peak_idx) - 1):
        a, b = peak_idx[i], peak_idx[i+1]
        if b <= a + 5:  # guard against tiny/invalid windows
            continue
        seg = signal[a:b]
        x_old = np.linspace(0, 1, num=seg.size)
        x_new = np.linspace(0, 1, num=points)
        resampled = np.interp(x_new, x_old, seg)
        cycles.append(resampled)
    if len(cycles) == 0:
        return np.empty((0, points))
    return np.vstack(cycles)

def summarize_cycles(cycles):
    """Return mean and std along the cycle dimension."""
    if cycles.size == 0:
        return None, None
    mu = cycles.mean(axis=0)
    sd = cycles.std(axis=0, ddof=1)
    return mu, sd

# ------------------------------------------------
# Pipeline: load → pick a foot → detect strides → normalize → aggregate
# ------------------------------------------------
def process_subject(file_path, foot="right"):
    left, right = load_subject_vgrf(file_path)
    sig = right if foot.lower().startswith("r") else left
    peaks = find_stride_peaks(sig, fs=FS, min_stride_s=0.8)
    # ensure we have at least a few strides
    if len(peaks) < 3:
        # try other foot if insufficient peaks
        alt = left if foot.lower().startswith("r") else right
        alt_peaks = find_stride_peaks(alt, fs=FS, min_stride_s=0.8)
        if len(alt_peaks) > len(peaks):
            sig, peaks = alt, alt_peaks
    cycles = normalize_cycles(sig, peaks, points=POINTS_PER_CYCLE)
    mu, sd = summarize_cycles(cycles)
    return mu, sd, cycles.shape[0]

# ------------------------------------------------
# Run for one PD and one HC subject
# ------------------------------------------------
pd_mu, pd_sd, pd_n = process_subject(PD_FILE, foot="right")
hc_mu, hc_sd, hc_n = process_subject(HC_FILE, foot="right")

# Safety check
if pd_mu is None or hc_mu is None:
    raise RuntimeError("Stride detection failed. Try switching feet or adjusting min_stride_s/thresholds.")

# ------------------------------------------------
# Plot: Mean normalized stride (0–100%) PD vs HC with ±1 SD
# (Single plot as requested.)
# ------------------------------------------------
x = np.linspace(0, 100, POINTS_PER_CYCLE)

plt.figure(figsize=(7.2, 4.2))
# HC
#plt.plot(x, hc_mu, label=f"Healthy Control (n={hc_n} strides)")
#plt.fill_between(x, hc_mu - hc_sd, hc_mu + hc_sd, alpha=0.25)
# PD
#plt.plot(x, pd_mu, label=f"Parkinson’s (n={pd_n} strides)")
#plt.fill_between(x, pd_mu - pd_sd, pd_mu + pd_sd, alpha=0.25)

plt.plot(x, hc_mu, color="#003865", lw=2.2, label="Healthy Control (n=66)")
plt.plot(x, pd_mu, color="#E57200", lw=2.2, label="Parkinson’s (n=111)")

plt.fill_between(x, hc_mu - hc_sd, hc_mu + hc_sd, color="#003865", alpha=0.15)
plt.fill_between(x, pd_mu - pd_sd, pd_mu + pd_sd, color="#E57200", alpha=0.25)

plt.ylim(0, 1500)
plt.xlim(0, 100)
plt.xlabel("Normalized Gait Cycle (%)")
plt.ylabel("Vertical Ground Reaction Force (N)")
plt.grid(True, linestyle='--', alpha=0.4)


#plt.xlabel("Normalized Stride (%)")
#plt.ylabel("vGRF (N)")
plt.title("Mean Normalized vGRF Stride — PD vs Healthy Control")
plt.legend()
#plt.grid(True, alpha=0.3)
plt.tight_layout()
# plt.savefig("mean_normalized_stride_PD_vs_HC.png", dpi=300, bbox_inches="tight", transparent=True)
plt.show()
```

```python
def load_demographics(data_dir: str) -> pd.DataFrame:
    """Read the PhysioNet demographics (xls/xlsx/html).
       Standardize to columns: subject_id (two-digit string), group (PD/CO), age, sex, label (PD=1, Control=0)."""
    # try excel first, fallback to html
    for name in ["demographics.xls", "demographics.xlsx"]:
        p = os.path.join(data_dir, name)
        if os.path.exists(p):
            df = pd.read_excel(p)
            break
    else:
        df = pd.read_html(os.path.join(data_dir, "demographics.html"))[0]

    # clean headers
    df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]

    # map columns (robust to variants)
    subj = next(c for c in ["code","subject","subject_id","id","subj","patient"] if c in df.columns)
    grp  = next(c for c in ["group","diagnosis","status","label"] if c in df.columns)
    age  = next((c for c in ["age","years"] if c in df.columns), None)
    sex  = next((c for c in ["sex","gender"] if c in df.columns), None)

    keep = [subj, grp] + [c for c in [age,sex] if c]
    df = df[keep].copy().rename(columns={subj:"subject_id", grp:"group"})
    if age: df = df.rename(columns={age:"age"})
    if sex: df = df.rename(columns={sex:"sex"})

    # subject ids in filenames are two-digit codes
    df["subject_id"] = df["subject_id"].astype(str).str.extract(r"(\d+)").fillna("0")
    df["subject_id"] = df["subject_id"].astype(int).astype(str).str.zfill(2)

    # label: PD=1, Control=0
    g = df["group"].astype(str).str.upper()
    df["label"] = np.where(g.str.contains("PD"), 1,
                   np.where(g.str.contains("CO")|g.str.contains("CONTROL")|g.str.contains("HEALTHY"), 0, np.nan))
    return df

demo = load_demographics(DATA_DIR)
print("demographics preview:")
display(demo.head())
print(f"subjects in demographics: {demo['subject_id'].nunique()}")
print("label counts:", demo["label"].value_counts(dropna=False).to_dict())
```

```python
def read_trial(path: str) -> pd.DataFrame:
    """Read one gait file as a DataFrame (tab-separated, no header)."""
    return pd.read_csv(path, sep="\t", header=None)

def parse_layout(df: pd.DataFrame):
    """Infer layout:
       - last two columns are totals
       - preceding columns are sensors, split equally into Left and Right."""
    ncols = df.shape[1]
    if ncols < 4:
        raise ValueError(f"file has only {ncols} columns (expected >=4).")
    # assume last two are totals
    SUM_A, SUM_B = ncols - 2, ncols - 1
    n_sensors = ncols - 2
    if n_sensors % 2 != 0:
        # if odd, drop the column just before totals (rare, defensive)
        n_sensors -= 1
        SUM_A, SUM_B = ncols - 3, ncols - 2
    per_foot = n_sensors // 2
    LEFT_COLS  = list(range(0, per_foot))
    RIGHT_COLS = list(range(per_foot, 2*per_foot))
    return LEFT_COLS, RIGHT_COLS, SUM_A, SUM_B

def assign_totals(df, LEFT_COLS, RIGHT_COLS, SUM_A, SUM_B):
    """Decide which total is LEFT vs RIGHT using correlation with per-foot means."""
    left_mean  = df.iloc[:, LEFT_COLS].mean(axis=1)
    right_mean = df.iloc[:, RIGHT_COLS].mean(axis=1)
    corrA_L = np.corrcoef(df.iloc[:, SUM_A], left_mean)[0,1]
    corrB_L = np.corrcoef(df.iloc[:, SUM_B], left_mean)[0,1]
    left_is_A = abs(corrA_L) >= abs(corrB_L)
    LEFT_SUM  = SUM_A if left_is_A else SUM_B
    RIGHT_SUM = SUM_B if left_is_A else SUM_A
    return LEFT_SUM, RIGHT_SUM

# test on one file to print the inferred layout once
ex_path = subjects["filepath"].iloc[0]
ex_df = read_trial(ex_path)
Lc, Rc, Sa, Sb = parse_layout(ex_df)
Ls, Rs = assign_totals(ex_df, Lc, Rc, Sa, Sb)
print(f"example file cols={ex_df.shape[1]} | L sensors={len(Lc)} | R sensors={len(Rc)} | LEFT_SUM={Ls} | RIGHT_SUM={Rs}")
```

```python
def trials_to_long(rows: pd.DataFrame, cols: list, tag: str) -> pd.DataFrame:
    """Convert a set of trials to TSFresh long format [id,time,kind,value] for the given columns."""
    out = []
    for _, r in rows.iterrows():
        df = read_trial(r.filepath)
        # re-derive layout per file (safe if shapes vary)
        Lc, Rc, Sa, Sb = parse_layout(df)
        use_cols = Lc if tag=="L" else Rc
        arr = df.iloc[:, use_cols].copy()
        arr = arr.reset_index().melt(id_vars="index", var_name="channel", value_name="value")
        arr = arr.rename(columns={"index":"time"})
        arr["id"] = r.trial_id
        arr["kind"] = f"{tag}_" + arr["channel"].astype(str)
        out.append(arr[["id","time","kind","value"]])
    return pd.concat(out, ignore_index=True)

right_long = trials_to_long(subjects, cols=None, tag="R")
left_long  = trials_to_long(subjects, cols=None, tag="L")
both_long  = pd.concat([right_long, left_long], ignore_index=True)

# order of IDs for consistent y
ids = subjects["trial_id"].tolist()
y   = subjects["label"].values.astype(float)  # not used for TSFresh, but for plots later
print("long tables ready:", len(right_long), len(left_long), len(both_long), "| trials:", len(ids))
```

```python
import os, re, glob
from pathlib import Path
import numpy as np
import pandas as pd

def _read_demographics_any(data_dir: str | Path) -> pd.DataFrame:
    """
    Tries xlsx/xls, then csv, then html. Raises RuntimeError if none found.
    """
    data_dir = Path(data_dir)
    # priority: xlsx/xls -> csv -> html
    cand = None
    for pat in ["demographics.xlsx", "demographics.xls", "demographics.csv", "demographics.html", "demographics.htm"]:
        f = data_dir / pat
        if f.exists():
            cand = f
            break
    if cand is None:
        # sometimes the file is named with caps or slightly different
        for f in data_dir.glob("demograph*.xls*"):
            cand = f; break
        if cand is None:
            for f in data_dir.glob("demograph*.csv"):
                cand = f; break
        if cand is None:
            for f in data_dir.glob("*.html"):
                # heuristically pick an html with 'demo' in name; otherwise first html
                if "demo" in f.name.lower() or "demograph" in f.name.lower():
                    cand = f; break
            if cand is None:
                hs = list(data_dir.glob("*.html"))
                if hs: cand = hs[0]

    if cand is None:
        raise RuntimeError(f"Could not find demographics file in {data_dir}")

    ext = cand.suffix.lower()
    if ext in [".xlsx", ".xls"]:
        df = pd.read_excel(cand)
    elif ext == ".csv":
        df = pd.read_csv(cand)
    else:
        # html -> take first table
        tables = pd.read_html(cand)
        if not tables:
            raise RuntimeError(f"Found {cand} but could not parse any table.")
        df = tables[0]

    # normalize column names: lower + spaces->underscore
    df.columns = [str(c).strip().lower().replace(" ", "_") for c in df.columns]

    # Fix duplicate columns on both pandas 1.x and 2.x
    if not df.columns.is_unique:
        # pandas >=2 usually dedups automatically; in case it doesn't:
        counts = {}
        new_cols = []
        for c in df.columns:
            if c not in counts:
                counts[c] = 0
                new_cols.append(c)
            else:
                counts[c] += 1
                new_cols.append(f"{c}.{counts[c]}")
        df.columns = new_cols

    return df


def load_demographics_clean(data_dir: str | Path) -> pd.DataFrame:
    """
    Returns a cleaned demographics DataFrame with:
        - subject_id: two-digit string '01','02',...
        - age (if present)
        - sex / gender (string if present)
        - label: 1=PD, 0=Control, NaN if unknown

    It is resilient to different column names and pandas versions.
    """
    df = _read_demographics_any(data_dir)

    # ----- choose a single subject-id column -----
    # candidates by priority
    cand_order = ["subject_id", "subject", "subj", "subjnum", "id", "patient"]
    lower_cols = {c.lower(): c for c in df.columns}
    subj_col = None
    for key in cand_order:
        if key in lower_cols:
            subj_col = lower_cols[key]
            break
    if subj_col is None:
        # heuristic: any column that contains 'subj' or ends with 'id'
        pats = [c for c in df.columns if ("subj" in c.lower()) or c.lower().endswith("_id") or c.lower()=="id"]
        subj_col = pats[0] if len(pats) else None

    if subj_col is None:
        raise RuntimeError(
            "Could not find a subject id column in demographics. "
            f"Columns seen: {list(df.columns)}"
        )

    s = df[subj_col]
    if isinstance(s, pd.DataFrame):   # just in case duplicate selection happened
        s = s.iloc[:, 0]
    s = s.astype(str)

    # IMPORTANT: expand=False -> Series (not DataFrame) so .str works
    subject_id = (
        s.str.extract(r"(\d+)", expand=False)
         .fillna("0")
         .astype(int).astype(str).str.zfill(2)
    )

    out = pd.DataFrame({"subject_id": subject_id})

    # ----- carry over useful columns if they exist -----
    # age
    if "age" in lower_cols:
        out["age"] = df[lower_cols["age"]]
    # sex/gender normalization
    sex_col = lower_cols.get("sex", None) or lower_cols.get("gender", None)
    if sex_col:
        sex_raw = df[sex_col].astype(str).str.strip().str.lower()
        # map a few common variants
        sex_map = {
            "m": "male", "male": "male",
            "f": "female", "female": "female",
            "0": "male", "1": "female"  # sometimes encoded; adjust if needed
        }
        out["sex"] = sex_raw.map(sex_map).fillna(sex_raw)
        # also keep gender as copy for safety if you like
        out["gender"] = out["sex"]

    # ----- derive binary label (if any column suggests group/diagnosis) -----
    # ----- derive binary label (if any column suggests group/diagnosis) -----
    label = pd.Series(np.nan, index=out.index, dtype="float64")

    # look for obvious group/diagnosis/status/label columns
    group_like = [c for c in df.columns if any(
        k in c.lower() for k in ["group", "diagnosis", "status", "label"]
    )]
    if group_like:
        # prefer 'group' if present
        col = None
        for key in ["group", "diagnosis", "status", "label"]:
            col = next((c for c in group_like if c.lower() == key), None)
            if col: break
        if col is None:
            col = group_like[0]

        g = df[col].astype(str).str.upper().str.strip()

        # expanded rules for more dataset variants
        label = np.select(
            [
                g.str.contains("PD") | g.str.contains("PARKINSON"),
                g.str.contains("CONTROL") | g.str.contains("HEALTHY") | g.str.contains("CO") | g.str.contains("CTRL")
            ],
            [1, 0],
            default=np.nan
        )
        label = pd.Series(label, index=out.index, dtype="float64")

    out["label"] = label

    # Final tidy: remove exact duplicates by subject_id (keep first)
    out = (out
           .drop_duplicates(subset=["subject_id"])
           .reset_index(drop=True))

    # Just to be explicit about dtypes
    out["subject_id"] = out["subject_id"].astype(str)

    return out
```
